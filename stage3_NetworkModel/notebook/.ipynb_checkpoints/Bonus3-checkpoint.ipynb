```python
import geopandas as gpd
```

```python
pathSHP = 'gis/osm/roads.shp'
```

```python
roadsSHP = gpd.read_file(pathSHP)
```

```python
roadsSHP.shape
```

```python
roads = {'N 1', 'N 102', 'N 2', 'N 204', 'N 207', 'R 170', 'R 203', 'R 220', 'R 241', 'Z 1005', 'Z 1031', 'Z 1042', 'Z 1044', 'Z 1065','Z 2013'}
```

```python
roadsSHP[roadsSHP['ref'].isin(roads)]['ref'].unique()
```

```python
roadsSHP['type'].unique()
```

```python
N1N2 = roadsSHP[roadsSHP['ref'].isin({"N 1",'N 2'})]
```

```python
otherRoads = roadsSHP[roadsSHP['ref'] != "N 1"]
```

```python
otherRoads = otherRoads[otherRoads['ref'] != 'N 2']
```

```python
otherRoads = otherRoads[otherRoads['ref'].notna()]
```

```python
nonres = otherRoads[otherRoads['type'] != "residential"]
```

```python
nonres = nonres[nonres['type'] != "tertiary"]
```

```python
nonres = nonres[nonres['type'] != "pedestrian"]
```

```python
nonres = nonres[nonres['type'] != "living_street"]
```

```python
nonres = nonres[nonres['type'] != "road"]
```

```python
N1N2.intersection(otherRoads)
```

```python
intersections = gpd.overlay(N1N2,nonres, how = 'intersection', keep_geom_type = False)
```

```python
intersections.shape
```

```python
intersections.to_csv("intersections.csv")
```
